```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
file_name = '/iris2.csv'
df = pd.read_csv(file_name)
df.head()
```

```python
X = {}
for i in range(4):
  X[i] = df.iloc[:, i:i+1]
```

```python
X[0]
```

```python
# 折れ線グラフ
plt.plot(X[0], X[1])
plt.plot(X[0], X[2])
plt.show()
```

```python
# 散布図
plt.scatter(X[0], X[1])
plt.scatter(X[0], X[2])
plt.show()
```

```python
# DataFrameから散布図の表示
df.plot.scatter('sepal length (cm)', 'petal width (cm)')
```

```python
sns.regplot(X[0], X[1])
plt.show()
```

```python
# ヒストグラム
plt.hist(df['sepal length (cm)'], bins=10) # 引数はcolumから選択
plt.xlabel('column1')
plt.ylabel('freq')
plt.title('histgram')
plt.show()
```

```python
sns.set()
sns.distplot(df['sepal length (cm)'], kde=True, bins=10, color='blue')
```

```python
# 複数のヒストグラムの描画
df_0 = X[1]
df_1 = X[2]
sns.distplot(df_0, kde=False, bins=20, color="r")
sns.distplot(df_1, kde=False, bins=20, color="b")
```

```python
# 複数のヒストグラムの描画
X1 = df['sepal length (cm)']
X2 = df['sepal width (cm)']
X3 = df['petal length (cm)']
plt.hist([X1, X2, X3], bins=10, color=['yellow', 'skyblue', 'blue'], label=['X1', 'X2', 'X3'], stacked=True)
plt.show()
```

```python
# 軸範囲指定, 凡例追加
plt.scatter(X[0], X[1], label= 'data A')
plt.scatter(X[0], X[2], label= 'data ')
plt.xlim(0, 160)
plt.ylim(0,10)
plt.legend(loc=2)
plt.show()
```

```python
# 複数のグラフの描画
plt.figure(figsize=(9,3))
plt,plt.subplot(131)
plt.scatter(X[0], X[1])
plt,plt.subplot(132)
plt.scatter(X[0], X[2])
plt,plt.subplot(133)
plt.scatter(X[0], X[3])
plt.show()
```

```python
x = df['sepal length (cm)']
plt.subplot(221)
plt.hist(x,bins=6)
plt.xlabel('sepal length (cm)')
plt.subplot(222)
x1 = df['sepal width (cm)']
plt.hist(x1,bins=6)
plt.xlabel('sepal width (cm)')
plt.subplot(223)
x2 = df['petal length (cm)']
plt.hist(x2,bins=6)
plt.xlabel('petal length (cm)')
plt.subplot(224)
x3 = df['petal width (cm)']
plt.hist(x3,bins=6)
plt.xlabel('petal width (cm)')
plt.show()
```
