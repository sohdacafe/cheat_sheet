```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
file_name = '/iris2.csv'
df = pd.read_csv(file_name)
df.head()
```

```python
X = {}
for i in range(4):
  X[i] = df.iloc[:, i:i+1]
```

```python
X[0]
```

```python
# 折れ線グラフ
plt.plot(X[0], X[1])
plt.plot(X[0], X[2])
plt.show()
```

```python
# 散布図
plt.scatter(X[0], X[1])
plt.scatter(X[0], X[2])
plt.show()
```

```python
# ヒストグラム
plt.hist(df['sepal length (cm)'], bins=10) # 引数はcolumから選択
plt.xlabel('column1')
plt.ylabel('freq')
plt.show()
```

```python
sns.set()
sns.distplot(df['sepal length (cm)'], kde=True, bins=10, color='blue')
# kernel density estimation(カーネル密度推定)
```

```python
# 複数のヒストグラムの描画
df_0 = X[1]
df_1 = X[2]
sns.distplot(df_0, kde=False, bins=20, color="r")
sns.distplot(df_1, kde=False, bins=20, color="b")
```
