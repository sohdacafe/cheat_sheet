```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
file_name = 'iris2.csv'
df = pd.read_csv(file_name)
df.head()
```

```python
df.columns
```

```python
df = df.drop('Unnamed: 0', axis=1)
```

```python
# 散布図の作成
sns.relplot(x='sepal length (cm)', y='petal length (cm)', data=df, hue='species')
# sns.relplot(x=df['sepal length (cm)'], y=df['petal length (cm)'], hue='species')
plt.show() 
```

```python
# スタイル指定(色とグリッドのついた背景スタイルを指定)
sns.set(style = 'darkgrid')
sns.relplot(x='sepal length (cm)', y='petal length (cm)', data=df, hue='species')
plt.show() 
```

```python
# 色変更：引数colorに色を指定
# マーカースタイル変更：引数styleにカテゴリを追加
# プロット色味変更：引数paletteに色を指定
```

```python
# カテゴリ別にグラフを分けて表示
sns.relplot(x='sepal length (cm)', y='petal length (cm)', data=df, style='species', hue='species', col='species')
plt.show() 
```

```python
# 特定の変量での散布図
sns.catplot(x='species', y='sepal length (cm)', data=df)
plt.show() 
```

```python
# 箱ひげ図(引数kindにboxを指定)
sns.catplot(x='species', y='sepal length (cm)', data=df, kind='box')
plt.show() 
```

```python
sns.violinplot(x='species', y='sepal length (cm)', data=df)
plt.show() 
```

```python
sns.distplot(df['sepal length (cm)'], hue='species')
plt.show() 
```

```python
# 回帰直線を含む散布図
sns.regplot(x='sepal length (cm)', y='petal length (cm)', data=df)
plt.show() 
```

```python
# ヒストグラムと合わせた散布図
sns.jointplot(x='sepal length (cm)', y='petal length (cm)', data=df)
plt.show() 
```

```python
# jiontplotに回帰直線と確率密度関数曲線を追加
sns.jointplot(x='sepal length (cm)', y='petal length (cm)', data=df, kind='reg')
plt.show() 
```

```python
# jiontplotをカテゴリ別に表示
sns.jointplot(x='sepal length (cm)', y='petal length (cm)', data=df, hue='species')
plt.show() 
```

```python
# ペアプロット
sns.pairplot(data=df, hue='species')
plt.show() 
```

```python
def reg_plot (df, x, y) :
    sns.regplot(x=x, y=y, data=df)
    plt.show() 
```

```python
reg_plot(df, 'sepal width (cm)', 'petal length (cm)')
```
