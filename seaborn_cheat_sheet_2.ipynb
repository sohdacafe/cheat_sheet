```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns  
```

```python
df = sns.load_dataset('tips')
df.head(3)
```

```python
# 散布図の作成
sns.set(style = 'darkgrid')
sns.scatterplot(x='total_bill', y='tip', data=df, hue='time')
plt.show()
```

```python
# グラフサイズの変更
plt.figure(figsize=(9,6))
sns.scatterplot(x='total_bill', y='tip', data=df, hue='time')
plt.show()
```

```python
# カテゴリ別に複数グラフの表示(relplot)
plt.figure(figsize=(9,6))
sns.relplot(x='total_bill', y='tip', data=df, col='time', kind='scatter', hue='sex')
plt.show()
```

```python
# relplotでのサイズ(比率)の変更(height, aspectで比率変更)
plt.figure(figsize=(9,6))
sns.relplot(x='total_bill', y='tip', data=df, col='time', kind='scatter', height=3, aspect=2)
plt.show()
```

```python
# 回帰直線付きグラフ表示(regplot)
sns.regplot(x='total_bill', y='tip', data=df)
plt.show()
```

```python
# 散布図にカテゴリ別回帰直線を追加表示(lmプロット)
sns.lmplot(x='total_bill', y='tip', data=df, hue='time')
plt.show()
# figureメソッドでグラフサイズ変更できない(height, aspectで比率変更)
```

```python
# ヒストグラム作成
plt.figure(figsize=(9,6))
sns.histplot(x='total_bill', data=df)
plt.show()
```

```python
# カテゴリ別ヒストグラム作成
plt.figure(figsize=(9,6))
sns.histplot(x='total_bill', data=df, hue='time', multiple='stack')
plt.show()
# multiple='stack'を指定しないと透明になる
```
